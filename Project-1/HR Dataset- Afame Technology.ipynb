# Binning the Age column

```python
bin_edges=[0,18,25,35,45,55,65,100]
bin_label=['0-18','19-25','26-35','36-45','46-55','56-65','Above 65+']
data['New_Age']=pd.cut(data['Age'],bins= bin_edges, labels=bin_label, right=False)

data['New_Age'].value_counts().sort_index()
```
